```python
from ultralytics import YOLO
import cv2
import numpy as np
import matplotlib.pyplot as plt

# โหลดโมเดล
model = YOLO("/content/runs/detect/train/weights/best.pt")

# ทดสอบกับภาพใหม่
image_path = "/content/pngtree-thai-coin-money-for-trading-exchange-expenses-bank-coin-photo-picture-image_5652518.jpg"
results = model.predict(image_path, conf=0.5)

# กำหนดสีกรอบของแต่ละเหรียญ
color_map = {
    "025": (0, 128, 255),   # สีส้ม
    "050": (255, 235, 0),   # สีฟ้า
    "1": (0, 252, 199),    # สีเขียว
    "2": (202, 250, 0),     # สีฟ้าอ่อน
    "5": (100,0,255),    # สีชมพู
    "10": (255,34,134),   # สีม่วง
}

# กำหนดสีตัวหนังสือตามประเภทเหรียญ
text_color_map = {
    "025": (0, 0, 0),  # สีดำ
    "1": (0, 0, 0),    # สีดำ
    "2": (0, 0, 0),    # สีดำ
    "050": (0, 0, 0),  # สีขาว
    "5": (255, 255, 255),    # สีขาว
    "10": (255, 255, 255),   # สีขาว
}

# โหลดภาพ
image = cv2.imread(image_path)

# วาดกรอบและข้อความ
for result in results:
    for box in result.boxes:
        x1, y1, x2, y2 = map(int, box.xyxy[0])  # พิกัดกรอบ
        label = result.names[int(box.cls[0])]   # ดึงชื่อคลาส
        conf = box.conf[0] * 100                # เปลี่ยนเป็นเปอร์เซ็นต์

        # เลือกสีกรอบ
        box_color = color_map.get(label, (255, 255, 255))  # ถ้าไม่มีให้ใช้สีขาว
        text_color = text_color_map.get(label, (255, 255, 255))  # สีตัวหนังสือ

        # ** วาดพื้นหลังสีดำใสในกรอบ **
        overlay = image.copy()
        cv2.rectangle(overlay, (x1, y1), (x2, y2), (0, 0, 0), -1)  # เติมพื้นหลังสีดำ
        alpha = 0.3  # ระดับความโปร่งใส
        image = cv2.addWeighted(overlay, alpha, image, 1 - alpha, 0)

        # ** วาดกรอบ (ไม่หนามาก) **
        cv2.rectangle(image, (x1, y1), (x2, y2), box_color, 2)

        # ** วาดพื้นหลัง Label สีเดียวกับกรอบ **
        font_scale = 0.6  # ปรับขนาดตัวหนังสือให้เล็กลง
        font_thickness = 2  # ลดความหนาของตัวหนังสือ
        label_text = f"{label} {int(conf)}%"

        label_size = cv2.getTextSize(label_text, cv2.FONT_HERSHEY_SIMPLEX, font_scale, font_thickness)[0]
        cv2.rectangle(image, (x1, y1 - label_size[1] - 13), (x1 + label_size[0] + 10, y1), box_color, -1)

        # ** วาดตัวหนังสือบน Label **
        cv2.putText(image, label_text, (x1 + 5, y1 - 5), cv2.FONT_HERSHEY_SIMPLEX, font_scale, text_color, font_thickness)

# แสดงภาพผลลัพธ์
img = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
plt.figure(figsize=(10, 6))
plt.imshow(img)
plt.axis("off")
plt.show()
```
