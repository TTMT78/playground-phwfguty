```python
from ultralytics import YOLO
import cv2
import numpy as np
import matplotlib.pyplot as plt

# à¹‚à¸«à¸¥à¸”à¹‚à¸¡à¹€à¸”à¸¥
model = YOLO("/content/runs/detect/train/weights/best.pt")

# à¸—à¸”à¸ªà¸­à¸šà¸à¸±à¸šà¸ à¸²à¸à¹ƒà¸«à¸¡à¹ˆ
image_path = "/content/pngtree-thai-coin-money-for-trading-exchange-expenses-bank-coin-photo-picture-image_5652518.jpg"
results = model.predict(image_path, conf=0.5)

# à¸à¸³à¸«à¸™à¸”à¸ªà¸µà¸à¸£à¸­à¸šà¸‚à¸­à¸‡à¹à¸•à¹ˆà¸¥à¸°à¹€à¸«à¸£à¸µà¸¢à¸
color_map = {
    "025": (0, 128, 255),   # à¸ªà¸µà¸ªà¹‰à¸¡
    "050": (255, 235, 0),   # à¸ªà¸µà¸Ÿà¹‰à¸²
    "1": (0, 252, 199),    # à¸ªà¸µà¹€à¸‚à¸µà¸¢à¸§
    "2": (202, 250, 0),     # à¸ªà¸µà¸Ÿà¹‰à¸²à¸­à¹ˆà¸­à¸™
    "5": (100,0,255),    # à¸ªà¸µà¸Šà¸¡à¸à¸¹
    "10": (255,34,134),   # à¸ªà¸µà¸¡à¹ˆà¸§à¸‡
}

# à¸à¸³à¸«à¸™à¸”à¸ªà¸µà¸•à¸±à¸§à¸«à¸™à¸±à¸‡à¸ªà¸·à¸­à¸•à¸²à¸¡à¸›à¸£à¸°à¹€à¸ à¸—à¹€à¸«à¸£à¸µà¸¢à¸
text_color_map = {
    "025": (0, 0, 0),  # à¸ªà¸µà¸”à¸³
    "1": (0, 0, 0),    # à¸ªà¸µà¸”à¸³
    "2": (0, 0, 0),    # à¸ªà¸µà¸”à¸³
    "050": (0, 0, 0),  # à¸ªà¸µà¸‚à¸²à¸§
    "5": (255, 255, 255),    # à¸ªà¸µà¸‚à¸²à¸§
    "10": (255, 255, 255),   # à¸ªà¸µà¸‚à¸²à¸§
}

# à¹‚à¸«à¸¥à¸”à¸ à¸²à¸
image = cv2.imread(image_path)

# à¸§à¸²à¸”à¸à¸£à¸­à¸šà¹à¸¥à¸°à¸‚à¹‰à¸­à¸„à¸§à¸²à¸¡
for result in results:
    for box in result.boxes:
        x1, y1, x2, y2 = map(int, box.xyxy[0])  # à¸à¸´à¸à¸±à¸”à¸à¸£à¸­à¸š
        label = result.names[int(box.cls[0])]   # à¸”à¸¶à¸‡à¸Šà¸·à¹ˆà¸­à¸„à¸¥à¸²à¸ª
        conf = box.conf[0] * 100                # à¹€à¸›à¸¥à¸µà¹ˆà¸¢à¸™à¹€à¸›à¹‡à¸™à¹€à¸›à¸­à¸£à¹Œà¹€à¸‹à¹‡à¸™à¸•à¹Œ

        # à¹€à¸¥à¸·à¸­à¸à¸ªà¸µà¸à¸£à¸­à¸š
        box_color = color_map.get(label, (255, 255, 255))  # à¸–à¹‰à¸²à¹„à¸¡à¹ˆà¸¡à¸µà¹ƒà¸«à¹‰à¹ƒà¸Šà¹‰à¸ªà¸µà¸‚à¸²à¸§
        text_color = text_color_map.get(label, (255, 255, 255))  # à¸ªà¸µà¸•à¸±à¸§à¸«à¸™à¸±à¸‡à¸ªà¸·à¸­

        # ** à¸§à¸²à¸”à¸à¸·à¹‰à¸™à¸«à¸¥à¸±à¸‡à¸ªà¸µà¸”à¸³à¹ƒà¸ªà¹ƒà¸™à¸à¸£à¸­à¸š **
        overlay = image.copy()
        cv2.rectangle(overlay, (x1, y1), (x2, y2), (0, 0, 0), -1)  # à¹€à¸•à¸´à¸¡à¸à¸·à¹‰à¸™à¸«à¸¥à¸±à¸‡à¸ªà¸µà¸”à¸³
        alpha = 0.3  # à¸£à¸°à¸”à¸±à¸šà¸„à¸§à¸²à¸¡à¹‚à¸›à¸£à¹ˆà¸‡à¹ƒà¸ª
        image = cv2.addWeighted(overlay, alpha, image, 1 - alpha, 0)

        # ** à¸§à¸²à¸”à¸à¸£à¸­à¸š (à¹„à¸¡à¹ˆà¸«à¸™à¸²à¸¡à¸²à¸) **
        cv2.rectangle(image, (x1, y1), (x2, y2), box_color, 2)

        # ** à¸§à¸²à¸”à¸à¸·à¹‰à¸™à¸«à¸¥à¸±à¸‡ Label à¸ªà¸µà¹€à¸”à¸µà¸¢à¸§à¸à¸±à¸šà¸à¸£à¸­à¸š **
        font_scale = 0.6  # à¸›à¸£à¸±à¸šà¸‚à¸™à¸²à¸”à¸•à¸±à¸§à¸«à¸™à¸±à¸‡à¸ªà¸·à¸­à¹ƒà¸«à¹‰à¹€à¸¥à¹‡à¸à¸¥à¸‡
        font_thickness = 2  # à¸¥à¸”à¸„à¸§à¸²à¸¡à¸«à¸™à¸²à¸‚à¸­à¸‡à¸•à¸±à¸§à¸«à¸™à¸±à¸‡à¸ªà¸·à¸­
        label_text = f"{label} {int(conf)}%"

        label_size = cv2.getTextSize(label_text, cv2.FONT_HERSHEY_SIMPLEX, font_scale, font_thickness)[0]
        cv2.rectangle(image, (x1, y1 - label_size[1] - 13), (x1 + label_size[0] + 10, y1), box_color, -1)

        # ** à¸§à¸²à¸”à¸•à¸±à¸§à¸«à¸™à¸±à¸‡à¸ªà¸·à¸­à¸šà¸™ Label **
        cv2.putText(image, label_text, (x1 + 5, y1 - 5), cv2.FONT_HERSHEY_SIMPLEX, font_scale, text_color, font_thickness)

# à¹à¸ªà¸”à¸‡à¸ à¸²à¸à¸œà¸¥à¸¥à¸±à¸à¸˜à¹Œ
img = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
plt.figure(figsize=(10, 6))
plt.imshow(img)
plt.axis("off")
plt.show()
```
